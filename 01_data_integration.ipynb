```python
import pandas as pd
import numpy as np

def convert_price(price):
    if pd.isna(price):
        return np.nan

    price = str(price).replace(',', '').strip()

    if price.lower() == 'price on request':
        return np.nan

    try:
        # Range values
        if '-' in price:
            if 'Cr' in price:
                nums = price.replace('Cr', '').split('-')
                nums = [float(x.strip()) for x in nums]
                return np.mean(nums) * 1e7

            elif 'L' in price:
                nums = price.replace('L', '').split('-')
                nums = [float(x.strip()) for x in nums]
                return np.mean(nums) * 1e5

        # Single values
        if 'Cr' in price:
            return float(price.replace('Cr', '').strip()) * 1e7

        if 'L' in price:
            return float(price.replace('L', '').strip()) * 1e5

        return float(price)

    except:
        return np.nan

all_df['price'] = all_df['price'].apply(convert_price)
```

```python
def convert_area(x):
    try:
        if isinstance(x, str):
            if '-' in x:
                nums = x.split('-')
                return (float(nums[0]) + float(nums[1]))/2
            else:
                return float(x)
        return x
    except:
        return np.nan

all_df['area_sqft'] = (
    all_df['area_sqft']
    .apply(convert_area)
)
```
